```python
for i in range(1, 525):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=<<api_key>>&language=en-US&page={}'.format(i))
    if response.status_code == 200:
        response_json = response.json()
        if 'results' in response_json:
            temp_df = pd.DataFrame(response_json['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
            df = pd.concat([df,temp_df],ignore_index=True)
        else:
            print(f"Warning: 'results' key not found in response for page {i}. Content: {response_json}")
            break
    else:
        print(f"Error: Request for page {i} failed with status code {response.status_code}")
        print(f"Response content: {response.text}")
        break
```
